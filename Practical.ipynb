```
from nltk.corpus import stopwords
from nltk.cluster.util import cosine_distance
import numpy as np
import networkx as nx
```

```
def read_file(file_name):
    file = open(file_name, "r")
    filedata = file.readlines()
    file_data = filedata[0].split(". ")
    sentences = []

    for sentence in file_data:
        print(sentence)
        sentences.append(sentence.replace("[^a-zA-Z]", " ").split(" "))
    sentences.pop() 
    
    return sentences
```

```
def similarityOfSentence(sentence1, sentence2, stopwords=None):
    if stopwords is None:
        stopwords = []
 
    sentence1 = [w.lower() for w in sentence1]
    sentence2 = [w.lower() for w in sentence2]
 
    all_words = list(set(sentence1 + sentence2))

    vector1 = [0] * len(all_words)
    vector2 = [0] * len(all_words)
    for w in sentence1:
        if w in stopwords:
            continue
        vector1[all_words.index(w)] += 1
    for w in sentence2:
        if w in stopwords:
            continue
        vector2[all_words.index(w)] += 1
    return 1 - cosine_distance(vector1, vector2)
```

```
def similarityMatrix(sentences, stop_words):
    similarity_matrix = np.zeros((len(sentences), len(sentences)))
 
    for idx1 in range(len(sentences)):
        for idx2 in range(len(sentences)):
            if idx1 == idx2: #in case both the sentences are same
                continue 
            similarity_matrix[idx1][idx2] = similarityOfSentence(sentences[idx1], sentences[idx2], stop_words)

    return similarity_matrix
```

```
def generateSummary(file_name, top_n=5):
    nltk.download("stopwords")
    stop_words = stopwords.words('english')
    summarize_text = []

    sentences =  read_file(file_name)

    similarityMatrixSentence = similarityMatrix(sentences, stop_words)

    similarity_graph = nx.from_numpy_array(similarityMatrixSentence)
    scores = nx.pagerank(similarity_graph)

    ranked_sentence = sorted(((scores[i],s) for i,s in enumerate(sentences)), reverse=True)    
    print("Indexes of top ranked_sentence order are ", ranked_sentence)    

    for i in range(top_n):
      summarize_text.append(" ".join(ranked_sentence[i][1]))

    print("Summarize Text: \n", ". ".join(summarize_text))
```
